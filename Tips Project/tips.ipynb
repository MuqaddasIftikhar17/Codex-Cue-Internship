# **Introduction** 📊💸

Dive into the world of dining and generosity with the Tips dataset! Originating from a vibrant restaurant environment 🍽️, this dataset provides fascinating insights into customer tipping behaviors. Whether you're a data enthusiast or just curious about how tips vary with different factors, this dataset is your gateway to exploring and understanding tipping dynamics.

---

# **Dataset Info** 📈

**Source**  
The Tips dataset is a classic included in the Seaborn library, a powerful tool for statistical data visualization built on top of Matplotlib 📉. This dataset gives you a taste of real-world tipping trends.

**Context**  
Imagine you're a data detective uncovering patterns from a real restaurant's tipping history. This dataset captures various features related to dining experiences, making it a perfect playground for data analysis.

---

# **Features 🔍**

The Tips dataset comes with a variety of columns that open up numerous avenues for exploration:

- **total_bill 💵**: The total amount of the bill (in dollars) before the tip.
- **tip 💸**: The amount of tip given (in dollars).
- **sex 🚻**: Gender of the bill payer ('Male' or 'Female').
- **smoker 🚬**: Whether the bill payer is a smoker ('Yes' or 'No').
- **day 📅**: Day of the week when the meal occurred (e.g., 'Thur', 'Fri', 'Sat', 'Sun').
- **time 🕒**: Time of day when the meal occurred ('Lunch' or 'Dinner').
- **size 👥**: Number of people in the dining party.

---

**Exploratory Data Analysis (EDA) 🔍**  
Uncover intriguing patterns with EDA! Discover how tips fluctuate by day of the week, the impact of smoking status, or how party size influences tipping behavior.

```python
# function to didplay histogram of total_bill column
def hist_plot(df, column, bins=50, alpha=0.5, kde=True):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], bins=bins, alpha=alpha, kde=kde)
    plt.title('HISTPLOT OF TOTAL_BILL')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    # show plot
    plt.show()

# function call
hist_plot(df, column='total_bill')    
```

```python
print('Mean =', df['total_bill'].mean()) # mean
print('Minimum Value =', df['total_bill'].min()) # minimum value
print('Maximum Value =', df['total_bill'].max()) # maximum value
print('Skewness =', df['total_bill'].skew()) # skewness
```

```python
# function to display histogram of tip column
def hist_plot(df, column, bins=50, alpha=0.5, kde=True):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], bins=bins, alpha=alpha, kde=kde)
    plt.title('DISTRIBUTION OF TIP COLUMN')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    # show plot
    plt.show()
    
# function call
hist_plot(df, column='tip')    
```
